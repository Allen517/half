```python
import numpy as np
import re
from collections import defaultdict
```

```python
def log_run_analy(filepath, keywords):
    recs = defaultdict(list)
    with open(filepath, 'r') as fin:
        for ln in fin:
            if 'Namespace' in ln:
                key = ''
                for kw in keywords:
                    pattern = re.compile(r'{}=(.*?),'.format(kw), re.DOTALL)
                    match = pattern.search(ln.strip())
                    if match:
                        key += '{}={};'.format(kw,match.group(1))
                key = key[:-1]
            if 'validation score:' in ln:
                pattern = re.compile(r'validation score: (.*)')
                match = pattern.search(ln.strip())
                if match:
                    recs[key].append(float(match.group(1)))
    return recs
```

```python
recs = log_run_analy('log/RUN.2019-04-25.log', ['gamma','eta'])

settings = []
vals = []
stds = []
for k,v in recs.items():
    settings.append(k)
    vals.append(np.mean(v))
    stds.append(np.std(v))
idx_srt_val = np.argsort(vals)[::-1]

for idx in idx_srt_val:
    print(settings[idx], vals[idx], stds[idx])
```
